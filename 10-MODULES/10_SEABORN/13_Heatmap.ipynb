# Heatmap 
    Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix. In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. Heatmap is also defined by the name of the shading matrix.

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
import random
np.random.seed(18)
data = np.random.randint(1,100,size = (5, 5))
print("The data to be plotted:\n")
print(data)
```

```python
hm = sn.heatmap(data = data)
plt.show()
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True) # vmin = start scale , vmax = end scale
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia')
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia',center = 70)
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia',center = 70,linewidths = 4)
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia',center = 70,linewidths = 4,linecolor = "k")
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia',center = 70,linewidths = 4,linecolor = "k",
                cbar = False,xticklabels = False,yticklabels = False)
```

```python
data2 = pd.DataFrame(data,columns=['c1','c2','c3','c4','c5'])
```

```python
data2.corr()
```

```python
sn.heatmap(data2.corr(), cmap="YlGnBu", annot=True)
```

```python
sn.heatmap(data2.corr(), cmap="YlGnBu", annot=True,mask = np.triu(np.ones_like(data2.corr())))
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,cmap='Wistia',center = 70,linewidths = 4)
for _, spine in hm.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(1)
    spine.set_linestyle("-")
```

```python
hm = sn.heatmap(data = data,vmin = 30,vmax = 80,annot = True,annot_kws={'size': 15},cmap='Wistia',center = 70,linewidths = 4)
```
