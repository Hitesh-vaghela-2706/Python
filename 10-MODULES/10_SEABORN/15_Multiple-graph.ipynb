# Multiple graphs Using Matplotlib
Matplotlib provides various functions for plotting subplots. Some of them are add_axes(), subplot(), and subplot2grid(). 

# Using add_axes() method

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
fig = plt.figure(figsize =(5, 4),facecolor='orange')
sns.set_palette("plasma")
ax1 = fig.add_axes([1, 0.5, 0.9, 0.9]) # ['x-axis','y-axis','width','hight']
graph()
sns.set_palette("tab10")
ax2 = fig.add_axes([0.1, 0.5, 0.7, 0.7]) 
graph()
plt.show()
```

# Using subplot() method

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
fig = plt.figure(figsize=(12,6),facecolor='orange')
sns.set_palette("plasma")
plt.subplot(1,2,1)
graph()
sns.set_palette("prism")
plt.subplot(1,2,2)
graph()
plt.show()
```

# Using subplot2grid() method

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
fig = plt.figure(figsize=(12,6),facecolor='orange')

sns.set_palette("plasma")
axes1 = plt.subplot2grid ( (7, 2), (0, 0), rowspan = 3, colspan = 2 )
graph()

sns.set_palette("prism")
axes2 = plt.subplot2grid ( (7, 2), (3, 0), rowspan = 2, colspan = 1)
graph()

sns.set_palette('gnuplot')
axes2 = plt.subplot2grid ( (7, 2), (4, 1), rowspan = 1, colspan = 1)
graph()

sns.set_palette('Wistia')
axes2 = plt.subplot2grid ( (7, 2), (5, 0), rowspan = 2, colspan = 1)
# axes2 = plt.subplot2grid ( (total row , total column ), ( draw on row no. , draw on coumn no. ),
# rowspan = no. of rows want to merge, colspan = no. of columns want to merge )
graph()

sns.set_palette('inferno')
axes3 = plt.subplot2grid ((7, 2), (6, 1), rowspan = 1, colspan = 1)
graph()
plt.show()
```

```python
fig = plt.figure(figsize=(20,20),facecolor='pink')
sns.set_context('poster')
sns.set_style("whitegrid")

sns.set_palette('magma')
g1 = plt.subplot2grid( ( 11,4 ),( 0,0 ) , rowspan = 3 , colspan = 2 )
sns.countplot(x="sex", data=data)


sns.set_palette('winter')
g2 = plt.subplot2grid( ( 11,4 ),( 0,2 ) , rowspan = 3 , colspan = 2 )
sns.lineplot(x="day", y="tip", data=data,ci= 20)


sns.set_palette('Pastel2')
g3 = plt.subplot2grid( ( 11,4 ),( 4,0 ) , rowspan = 3 , colspan = 2 )
sns.scatterplot(x="total_bill", y="tip",hue='smoker', data=data)


sns.set_palette('vlag')
g4 = plt.subplot2grid( ( 11,4 ),( 4,2 ) , rowspan = 3 , colspan = 2 )
sns.barplot(x="day", y="tip", data=data)


sns.set_palette('copper')
g5 = plt.subplot2grid( ( 11,4 ),( 8,0 ) , rowspan = 3 , colspan = 2 )
sns.violinplot(x="day", y="tip", data=data)


sns.set_palette('Wistia')
g6 = plt.subplot2grid( ( 11,4 ),( 8,2 ) , rowspan = 3 , colspan = 2 )
sns.boxplot(x="sex", y="tip", data=data)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'Date': ['2019-10-01', '2019-11-01','2019-12-01','2020-01-01','2020-02-01', '2020-03-01',
                            '2020-04-01', '2020-05-01','2020-06-01'],
                   'Col_1': [34, 43, 14, 15, 15,14, 31, 25, 62],
                   'Col_2': [52, 66, 78, 15, 15, 5, 25, 25, 86],
                   'Col_3': [13, 73, 82, 58, 52,87, 26, 5, 56],
                   'Col_4': [44, 75, 26, 15, 15,14, 54, 25, 24]})

plt.figure(figsize=(12,12))

g1 = plt.subplot2grid( (7,2),(0,0),rowspan=3,colspan=1)
sns.lineplot( x = "Date", y = "Col_1",color = 'r', data = df)
plt.xticks(rotation = 25)

g2 = plt.subplot2grid( (7,2),(0,1),rowspan=3,colspan=1)
sns.lineplot( x = "Date", y = "Col_2",color = 'g', data = df)
plt.xticks(rotation = 25)

g3 = plt.subplot2grid( (7,2),(4,0),rowspan=3,colspan=1)
sns.lineplot( x = "Date", y = "Col_3",color = 'b', data = df)
plt.xticks(rotation = 25)

g4 = plt.subplot2grid( (7,2),(4,1),rowspan=3,colspan=1)
sns.lineplot( x = "Date", y = "Col_4",color = 'y', data = df)
plt.xticks(rotation = 25)

plt.show()
```
