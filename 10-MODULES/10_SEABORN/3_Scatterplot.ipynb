# Scatter Plot

```python
import seaborn
seaborn.set(style='whitegrid')
fmri = seaborn.load_dataset("fmri")
fmri
```

```python
seaborn.scatterplot(x="timepoint",y="signal",data=fmri)
```

```python
seaborn.scatterplot(x="timepoint",y="signal",hue="region",data=fmri)
```

```python
tip = seaborn.load_dataset('tips')
tip
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip)
```

```python
seaborn.scatterplot(x='day', y='tip', data= tip, marker = 'D')
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip, hue='time',size = "time")
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip, hue="time", style="time")
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip, hue='time', palette='plasma')
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip, hue='day',size = "day", palette='Blues',marker = 'D',alpha = 0.7)
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip ,hue='size', size = "size")
```

```python
seaborn.scatterplot(x='day', y='tip', data=tip, alpha = 0.1)
```
