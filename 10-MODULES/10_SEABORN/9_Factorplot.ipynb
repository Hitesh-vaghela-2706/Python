# Factorplot

seaborn.factorplot() method is used to draw a categorical plot onto a FacetGrid.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df
```

```python
sns.factorplot(x ='day', y ='tip', data = df,hue='sex',row='smoker',col='time',ci=100,
               order=['Sat','Fri','Thur','Sun'], hue_order=['Male','Female'], row_order=['No','Yes'], 
               col_order=['Dinner','Lunch'],height=5, aspect=2, orient='v',color='yellow', palette='Set3_r',
               legend=True,sharex=False, sharey=False,kind='bar'
              )
# this is a depriciated so instead use catplot
```

```python
sns.catplot(x ='day', y ='tip', data = df,hue='sex',row='smoker',col='time',ci=100,
               order=['Sat','Fri','Thur','Sun'], hue_order=['Male','Female'], row_order=['No','Yes'], 
               col_order=['Dinner','Lunch'],height=5, aspect=2, orient='v',color='yellow', palette='Set3_r',
               legend=True,sharex=False, sharey=False,kind='bar'
              )
# kind='bar',kind="strip",kind="swarm",kind="box",kind="boxen",kind="violin",kind="count",kind="point"
```
