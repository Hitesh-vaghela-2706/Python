```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.catplot(x='day',y='tip',data=data)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',hue_order=["Female","Male"])
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',hue_order=["Female","Male"],row='time')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',hue_order=["Female","Male"],col='time')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',hue_order=["Female","Male"],row='smoker',col='time')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',hue_order=["Female","Male"],row='smoker',col='time',row_order=['No','Yes'], 
               col_order=['Dinner','Lunch'])
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3, orient='v')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3, orient='v',color='yellow')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3, orient='v',color='yellow',palette='Set3_r',kind='bar')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3, orient='v',color='yellow',palette='Set2_r',
            kind='bar',order=['Sat','Fri','Thur','Sun'],ci=40)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3,palette='Set2_r',kind='bar',ci=40,legend=False)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',height=3, aspect=3,palette='Set2_r',kind='bar',ci=40,sharey=True,sharex=False)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',palette='Set1_r',kind='strip')
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',palette='Set1_r',kind="swarm",height=6)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',palette='Set1_r',kind="point",height=6)
```

```python
sns.catplot(x='day',y = None,data=data,hue='sex',palette='Set1_r',kind="count")
```

```python
sns.catplot(x='day',y='tip',data=data,palette='Set1_r',kind="boxen")
```

```python
sns.catplot(x='day',y='tip',data=data,palette='Set1_r',kind="box")
```

```python
sns.catplot(x='day',y='tip',data=data,palette='Set1_r',kind="violin")
```
