# Violin Plot
A violin plot plays a similar activity that is pursued through whisker or box plot do. As it shows several quantitative data
across one or more categorical variables. It can be an effective and attractive way to show multiple data at several units. A
“wide-form” Data Frame helps to maintain each numeric column which can be plotted on the graph. It is possible to use NumPy or Python objects, but pandas objects are preferable because the associated names will be used to annotate the axes.

```python
import seaborn as sns
sns.set(style = 'whitegrid')
tips = sns.load_dataset('tips')
tips
```

```python
sns.violinplot(x ='day', y ='tip', data = tips)
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex')
```

```python
sns.violinplot(x ='day', y ='tip',hue='sex', data = tips,style='day')
```

```python
sns.violinplot(x ='day', y ='tip',hue='sex', data = tips,size='time')
```

```python
sns.violinplot(x="total_bill",data=tips)
```

```python
sns.violinplot(y="total_bill",data=tips,inner="quart")
```

```python
sns.violinplot(x ='day', y ='tip', data = tips, linewidth = 1)
```

```python

sns.violinplot(x ='day', y ='tip', data = tips, linewidth = 1,palette="Set2", dodge=True)
```

```python
sns.violinplot(x ='day', y ='tip', data = tips, linewidth = 1,palette="Set2",order=["Sat", "Fri",'Sun','Thur'])
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1,palette="pastel",order=["Sat", "Fri",'Sun','Thur'],
               hue_order=["Female", "Male"])
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1,palette="pastel",order=["Sat", "Fri",'Sun','Thur'],
               hue_order=["Female", "Male"],saturation =0.3)
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1, color = "Yellow",
               order=["Sat", "Fri",'Sun','Thur'],hue_order=["Female", "Male"],
               saturation =0.3,style='day',dodge=True) # palette="pastel"
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1, color = "Yellow",
               order=["Sat", "Fri",'Sun','Thur'],hue_order=["Female", "Male"],
               saturation =0.3,style='day',dodge=True,split=True) # palette="pastel"
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1,palette="pastel",
               order=["Sat", "Fri",'Sun','Thur'],hue_order=["Female", "Male"],
               saturation =3,style='day',dodge=True,split=True,scale="count")
```

```python
sns.violinplot(x ='day', y ='tip', data = tips,hue='sex', linewidth = 1,palette="pastel",
               order=["Sat", "Fri",'Sun','Thur'],hue_order=["Female", "Male"],
               saturation =3,style='day',dodge=True,split=True,scale="count",inner="quart",orient='v',
               gridsize=100,width=0.8)
```
