# Kernel Density Estimate (KDE) Plot 

    Kdeplot allows us to estimate the probability density function of the continuous or non-parametric from our data set curve in one or more dimensions it means we can create plot a single graph for multiple samples which helps in more efficient data visualization.

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x= np.random.randn(200)
y = np.random.randn(200)
```

```python
sns.kdeplot(x = x)
```

```python
sns.kdeplot(x = y)
```

```python
sns.kdeplot(x = y,fill=True)
```

```python
sns.kdeplot(x = x, shade = True)
```

```python
sns.kdeplot(x = x, shade = True , color = "Green")
```

```python
sns.kdeplot(y = x,color = "Green")
```

```python
sns.kdeplot(y = x,color = "salmon",shade = True)
```

```python
sns.kdeplot(x = x,y = y,shade=True,color='yellow')
```

```python
sns.kdeplot(x = x,y = y,shade=True,palette = 'plasma')
```

```python
sns.kdeplot(x = x,y = y,shade=True,palette = 'plasma',cmap = "winter_r")
```

```python
tips = sns.load_dataset('tips')
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',cmap = "winter_r",data=tips)
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex') # ignore cmap if using hue
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'])
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=True)
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=True,
           multiple = "layer")
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=True,
           multiple = "stack")
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=False,
           multiple = "fill")
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=False,
           multiple = "fill",levels=5,thresh=0.2)
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=False,
           multiple = "fill",levels=5,thresh=0.05,gridsize=100,cut=6)
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=False,
           multiple = "fill",levels=5,thresh=0.05,gridsize=100,cut=6,bw_adjust=6)
```

```python
sns.kdeplot(x = 'tip',y = 'size',shade=True,palette = 'plasma',data=tips,hue='sex',hue_order = ['Male','Female'],cbar=False,
           multiple = "fill",levels=5,thresh=0.05,gridsize=100,cut=6,bw_adjust=4,bw_method='scott')
```
