# LINEPLOT
A single line plot presents data on x-y axis using a line joining datapoints.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
A = np.random.randint(1,30,20)
np.random.seed(23)
B = np.random.randint(50,260,20)
C = np.random.choice(['old','new','midium'],20)
data = pd.DataFrame({'year':A,'Speed':B,'Model':C})
data
```

```python
plt.figure(figsize=(10,5),facecolor='white')
sns.set_style('darkgrid')
sns.set_context('paper',font_scale=1.5)
sns.set_palette('ocean')
sns.lineplot(x = 'year',y = 'Speed',data = data)
plt.xlabel('X-axis years')
plt.ylabel('Y-axis Speed')
plt.title('Lineplot using seaborn')
plt.show()
```

```python
sns.lineplot(x = 'year',y = 'Speed',hue='Model', data = data)
```

```python
sns.lineplot(x = 'year',y = 'Speed',hue='Model', size='Model' , data = data)
```

```python
sns.lineplot(x = 'year',y = 'Speed',hue='Model',style='Model' ,data = data,ms = 110,legend=True)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.lineplot(x = 'year',y = 'Speed',palette="plasma",data = data,estimator=sum)
```

```python
sns.lineplot(x = 'year',y = 'Speed',palette="plasma",data = data,estimator=sum,ci=95)
```

```python
sns.lineplot(x = 'year',y = 'Speed',palette="Reds",data = data,sort=True)
```

```python
sns.lineplot(x = 'year',y = 'Speed',palette="Reds",data = data,err_style = "band")
```

```python
sns.lineplot(x = 'year',y = 'Speed',palette="Reds",data = data,err_style = "bars")
```

```python
tips_df=sns.load_dataset('tips')
sns.lineplot(x='size',y='total_bill',data=tips_df,ci=30)
```

```python
sns.lineplot(x='size',y='total_bill',data=tips_df,ci=30)
plt.xlim(2)
plt.ylim(15)
plt.show()
```

```python
sns.set_style("whitegrid",rc=None) # other style 'dark','darkgrid','white','ticks'
sns.set_palette('plasma')
sns.set_context("paper",font_scale=2,rc=None) # other context "poster","talk","notebook"
sns.lineplot(x='size',y='total_bill',data=tips_df,ci=30)
```

```python
sns.set_theme(context='paper',style='whitegrid',palette='plasma',font='sans-serif',font_scale=2,rc=None)
sns.lineplot(x='size',y='total_bill',data=tips_df,ci=30)
```
