 # Strip plot
    It is used to draw a scatter plot based on the category.
## Syntax: 
    seaborn.stripplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, 
                    orient=None, color=None, palette=None, size=5, edgecolor=’gray’, linewidth=0, ax=None, **kwargs)

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips
```

```python
sns.set(style = 'darkgrid')
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x = "total_bill" , data = tips)
```

```python
sns.stripplot(y = "total_bill" , data = tips)
```

    jitter can be used to provide displacements along the horizontal axis, which is useful when there are large clusters of data points. You can specify the amount of jitter, or just use True for a good default.

```python
sns.stripplot(x="sex", y="total_bill", data=tips, jitter = 0.1)
```

```python
sns.stripplot(x="sex", y="total_bill", data=tips,linewidth=0.9)
```

```python
sns.stripplot(x="sex", y="total_bill", data=tips,edgecolor='gray')
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker' ,data=tips,edgecolor='gray')
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker' ,data=tips,edgecolor='gray',palette="Set2")
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker' ,data=tips,edgecolor='green',linewidth=2,palette="Set2")
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker' ,data=tips,edgecolor='green',linewidth=2,color='yellow')
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker',data=tips,edgecolor='blue',linewidth=2,color='yellow',size=17)
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker',data=tips,edgecolor='blue',linewidth=2,color='yellow',size=17,
             marker="*", alpha=0.8)
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker',data=tips,edgecolor='salmon',linewidth=1,
              palette='Greens',size=17,marker="*", alpha=0.8,order=['Female','Male'])
```

```python
sns.stripplot(x="sex", y="total_bill",hue='smoker',data=tips,edgecolor='white',linewidth=0.4,palette='ocean',size=17,
             marker="*", alpha=0.8,order=['Female','Male'],hue_order=['No','Yes'])
```
