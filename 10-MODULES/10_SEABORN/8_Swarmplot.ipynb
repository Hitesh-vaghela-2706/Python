# Swarm Plot
    Seaborn swarmplot is probably similar to stripplot, only the points are adjusted so it won’t get overlap to each other as it helps to represent the better representation of the distribution of values. A swarm plot can be drawn on its own, but it is also a good complement to a box, preferable because the associated names will be used to annotate the axes. This type of plot sometimes known as “beeswarm”.

# Syntax :
    seaborn.swarmplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None,
        color=None,palette=None, size=5, edgecolor=’gray’, linewidth=0, ax=None, **kwargs)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
tips
```

```python
sns.swarmplot(x='tip', data=tips)
```

```python
sns.swarmplot(y ='tip', data=tips)
```

```python
sns.swarmplot(x='day', y='tip', data=tips,size = 4)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='tip', y='day', data=tips,size = 6)
```

```python
sns.swarmplot(x='day', y='tip', data=tips,size = 4,hue='time')
```

```python
sns.swarmplot(x='day', y='tip', data=tips,size = 4,hue='time',linewidth=0.8)
```

```python
sns.swarmplot(x='day', y='tip', data=tips,size = 4,hue='time',linewidth=0.8,edgecolor='pink')
```

```python
sns.swarmplot(x='day', y='tip', data=tips,size = 4,hue='time',linewidth=0.8,edgecolor='pink',palette='plasma')
```

```python
plt.figure(figsize=(16,8),facecolor='yellow')
sns.swarmplot(x='day', y='tip', data=tips,size = 10,hue='time',linewidth=2,edgecolor='pink',palette='Set2',
             order=['Sat','Sun','Fri','Thur'],hue_order=['Lunch','Dinner'],marker="D",alpha = 0.9)
```

```python
plt.figure(figsize=(16,8),facecolor='salmon')
sns.swarmplot(x='day', y='tip', data=tips,size = 10,hue='time',linewidth=2,edgecolor='g',color='yellow',
             order=['Sat','Sun','Fri','Thur'],hue_order=['Lunch','Dinner'],marker="D",alpha = 0.9)
plt.show()
```
