```python
B = list(data["date"])
print(B)
D = []
for i in B:
    C= i.split("-")
    D.append(C)
# print(D)
day = []
month = []
year = []

for i in range(len(D)):
    for j in range(len(D[i])):
        if(j == 0 ):
            V = D[i][j].replace("'","")
            day.append(int(V))
        if(j == 1 ):
            month.append(int(D[i][j]))
        if(j == 2):
            V = D[i][j].replace("'","")
            year.append(int(V))
# print("\n\n",day,end='')
# print("\n\n",month,end='')
# print("\n\n",year,end='')

X = pd.DataFrame(day,columns=['Day'])
Y = pd.DataFrame(month,columns=['Month'])
Z = pd.DataFrame(year,columns=['Year'])
W = pd.concat([X,Y,Z],axis=1)
# print("\n\n",W)
A = pd.merge(data,W ,left_index=True,right_index=True)
A
```
