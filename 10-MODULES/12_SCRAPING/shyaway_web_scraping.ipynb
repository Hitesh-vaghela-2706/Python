```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
#Declaring the headers 
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',"Accept-Language": "en-US,en;q=0.5"}
```

```python
#declaring the list of empty variables, So that we can append the data overall

name1 = []
discounted_price1 = []
original_price1 = []
discount1 = []
```

```python
#creating an array of values and passing it in the url for dynamic webpages
pages = np.arange(1,51)
```

```python
pages
```

```python
#the whole core of the script
for page in pages: 
    page = requests.get("https://www.shyaway.com/bra-online/?p=" + str(page) + "")
    soup = BeautifulSoup(page.text, 'html.parser')
    bras = soup.find_all(class_="item product product-item")
    for bra in bras:
        detail = bra.find('div',class_="product details product-item-details")
        try:
            name = detail.find('a',class_="product-name").text
        except:
            name = np.nan
        try:
            discounted_price = detail.find('span',class_="special-price").text
        except:
            discounted_price = np.nan
        try:
            original_price = detail.find('span',class_="old-price").text
        except:
            original_price = np.nan
        try:
            discount = detail.find('span',class_="saved-price").text
        except:
            discount = np.nan
        name1.append(name)
        discounted_price1.append(discounted_price)
        original_price1.append(original_price)
        discount1.append(discount)
```

```python
D = {
    'name' : name1 ,
    'discounted_price' : discounted_price1,
    'original_price' : original_price1,
    'discount' : discount1
}
```

```python
import pandas as pd
df = pd.DataFrame(D)
df
```
