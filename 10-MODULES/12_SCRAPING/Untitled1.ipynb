```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://www.shyaway.com/bra-online/?p=2"
```

```python
Headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'} 
```

```python
Page = requests.get(URL,Headers)
```

```python
Page
```

```python
Soup = BeautifulSoup(Page.content,'html.parser')
```

```python
Soup
```

```python
bras = Soup.find_all(class_="item product product-item")
# bras = Soup.find_all(class_="products wrapper grid products-grid")
```

```python
list1 = []
for bra in bras:
    list2 = []
    detail = bra.find('div',class_="product details product-item-details")
    name = detail.find('a',class_="product-name").text
    discounted_price = detail.find('span',class_="special-price").text
    original_price = detail.find('span',class_="old-price").text
    discount = detail.find('span',class_="saved-price").text
    list2.append(name)
    list2.append(discounted_price)
    list2.append(original_price)
    list2.append(discount)
    list1.append(list2)
```
