```python
import requests
from bs4 import BeautifulSoup
```

```python
page = [i for i in range(1,50)]
```

```python
K = 1
for i in page:
    url = f"https://www.shyaway.com/bra-online/?p={i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    bras = soup.find_all(class_="item product product-item")
    for bra in bras:
        image = bra.find("img", class_="product-image-photo") 
        image_url = image["src"]
        response = requests.get(image_url)
        file_path = f"D:\Bra\image{K}.jpg"
        with open(file_path, "wb") as file:
            file.write(response.content)
        K += 1
```
