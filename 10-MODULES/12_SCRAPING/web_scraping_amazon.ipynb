```python
from bs4 import BeautifulSoup
import requests
```

```python
# URL = "https://www.amazon.in/s?k=bra&crid=2LXN2IO33OKIE&sprefix=bra%2Caps%2C362&ref=nb_sb_noss_1"
URL = "https://www.amazon.in/s?k=transparent+bra+for+women&crid=VPYKQYY6Y8H0&sprefix=transparent+bra+for+women%2Caps%2C332&ref=nb_sb_ss_ts-doa-p_1_25"
```

```python
Headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'} 
```

```python
Page = requests.get(URL,Headers)
```

```python
Soup = BeautifulSoup(Page.content,'html.parser')
```

```python
Soup
```

```python
bras = Soup.find_all(class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro")
```

```python
list1 = []
for bra in bras:
    list2 = []
    brand_bra = bra.find('span' , class_="a-size-base-plus a-color-base").text
    description_bra = bra.find('span',class_="a-size-base-plus a-color-base a-text-normal").text
#     rating_bra =  bra.find('span',class_="a-icon-alt").text
    price_bra = bra.find('span', class_="a-offscreen").text
    list2.append(brand_bra)
    list2.append(description_bra)       
    list2.append(price_bra)
    list1.append(list2)
```

```python
len(list1)
```

```python
name = []
for i in range(len(list1)):
    name.append(list1[i][0])
name
```

```python
description = []
for i in range(len(list1)):
    description.append(list1[i][1])
description
```

```python
price = []
for i in range(len(list1)):
    price.append(list1[i][2])
price
```

```python
title = []
for i in range(len(list1)):
    title.append(f"Bra{i}")
title
```
