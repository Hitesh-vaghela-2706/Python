```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from( [ (1,2), (1,3), (3,4), (1,5), (4,0), (5,0), (2,5), (2,3), (2,0), (0,3)] )
```

```python
nx.draw(G,with_labels=True, node_color="yellow" , node_size=1500 , font_weight= "bold" , edge_color="blue")
```
