```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_data_set.csv')
```

```python
x=df['Property_Area'].unique() 
color=['b','y','r']

for i,j in zip(x,color): 
    df1=df.loc[df['Property_Area']==i] 
    plt.scatter(df1['Loan_ID'],df1['LoanAmount'],c=j) 


plt.legend(x)

plt.show()
```
