```python
import pandas as pd
import re
from io import StringIO
```

```python
data = ('Gender,Age,Amount,Mobile,Email\n'
         'male,19,19000,8000618586,xyz@gmail.com\n'
         'male,20,20000,98659865598,xyz@gmail.com\n'
         'male,23,3000,9909224943,xyz@yahoo.com\n'
         'male,52,20000,9712109704,xyz@gmail.com\n'
         'female,19,19000,854678546785,xyz@gmail.com\n'
         'male,19,20000,2546875946,xyz@yahoo.com\n'
         'female,23,1000,9112667856,xyz@gmail.com\n'
         'male,52,25000,1235486795,xyz@gmail.com\n'
         'female,20,16000,7041982574,xyz@gmail.c\n'
         'male,23,13000,7874199585,xyz@gmail.com\n'
       )
```

```python
df = pd.read_csv(StringIO(data))
regex = '^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'
```

```python
def check_email(email):
    if(re.fullmatch(regex, email)):
        return True
    else:
        return False
df['is_valid_email'] = df['Email'].apply( lambda x : check_email(str(x)))
```

```python
Pattern = ("(0/91)?[5-9][0-9]{9}")
def check_number(mobile):
    if(re.fullmatch(Pattern ,mobile)):
        return True
    else:
        return False
df['is_valid_phone'] = df['Mobile'].apply(lambda x:check_number(str(x)))
```
